# PART-A 

**Airline A is currently flying a large international network. Their flight operations are shown in the tab "Operations". Various aircraft characteristics are listed in tab "AC characteristics". They would like to know the following:**



**1. What is the total cost by aircraft type for the year? (Suggested tools: Any of Python/SQL )**
   * Note:Total cost = Hours flown * Costs/hour
   
   
**2. Which aircraft type  has the lowest cost per seat per km flown?**


#### In order to get the desired columns we have implemented Data Manipulation using Pandas which is an efficient manipulation tool
#### Here we have copied the AC characteristics by removing the unwanted entries on a new sheet named "Formmated AC characteristics" for easy data manipulation

### Characteristics

```python
# AC CHARACTERISTICS
import pandas as pd
df = pd.read_excel('Data Science Case study Vindiata.xls',sheet_name = 'Formatted AC characteristics')
df.head()
```

### Operations
#### Here we have copied the operations by removing the unwanted entries on a new sheet named "Formmated Operations" for easy data manipulation

```python
oprtn_sh = pd.read_excel('Data Science Case study Vindiata.xls', sheet_name = 'Formatted Operations')
columns = ['Jan-14','Feb-14','Mar-14','Apr-14','May-14','Jun-14','Jul-14','Aug-14','Sep-14','Oct-14','Nov-14','Dec-14']
oprtn_sh.rename(columns=dict(zip(oprtn_sh.columns[2:], columns)), inplace=True)
oprtn_sh
```

### City Pairs
#### Here we have copied the city pairs by removing the unwanted entries on a new sheet named "Formatted City pairs" for easy data manipulation

```python
city_pairs = pd.read_excel('Data Science Case study Vindiata.xls', sheet_name = 'Formatted City pairs')
city_pairs.rename(columns = {'Pass. Demand \n(per day)':'Pass. Demand (per day)'}, inplace = True)
city_pairs
```

```python
# Creating a new column "Total hours flown"
oprtn_sh['Total hours flown'] = oprtn_sh.iloc[:,2:].sum(axis = 1)
oprtn_sh.head()
```

```python
# Calculating the sum of each type of aircraft and merging it with the Aircraft characteristics
oprtn_copy = oprtn_sh.copy()
oprtn_copy = oprtn_copy.groupby(['Aircraft Type']).sum()
operations = df.merge(oprtn_copy, how = 'left', on = 'Aircraft Type')
operations
```

### TOTAL COST ACCORDING TO AIRCRAFT TYPE
**Total Cost = Total Hours Flown * Cost/hour**

```python
operations['Total Cost in USD'] = operations['Total hours flown'] * operations['Costs per flight hour']
operations[['Aircraft Type','Total Cost in USD']]
```

### Cost per seat per Km(By Aircraft Type)
#### Cost per seat per km means that we need to calculate the cost of one seat in a particular type of  Aircraft for each kilometer flown.
* For that we need to divide the total cost of that type of aircraft with the product of Number of seats, total hours flown and the speed of the aircraft


**Cost per seat per km = Total cost/(No. of seats x Total hours x Avg. speed)**


```python
operations['Cost per seat per km(USD)'] = operations['Total Cost in USD']/(operations['Number of Seats'] * operations['Total hours flown'] * operations['Ave. Speed (km/h)'])
# Sorting should be on the Cost per seat per km(USD) 
operations[['Aircraft Type','Cost per seat per km(USD)']].sort_values(by = ['Cost per seat per km(USD)'])
```

##### OUTCOME : Aircraft A330 is the most economical aircraft type which is the least expensive and costs 0.033333$ per seat per km flown


# PART-B

**Airline B is a startup airline and is considering flying on the city pairs as shown in tab "City Pairs". They would like to know:**

**1. Which aircraft types are best suited for their operation?  (Suggested tools: Any of Python/SQL )**

**Note:	This is an optimization between 3 factors: Range, Passenger demand, costs. Hence the aircraft which is capable of flying between 2 cities (range), has enough seats to meet the demand, and has the lowest cost will win. Please also note that the passenger demand is per day, hence an aircraft can make several trips between 2 cities to meet that demand. 1 aircraft type will always be most suitable for a given city pair. Hence resultant fleet would be a mix of different aircraft types. In case you are unable to get to the final solution, please lay out the step-by-step approach to be used.**													


```python
# Removal of irrelavant columns necessary for evaluation
operations.drop(operations.iloc[:, 5:17], axis = 1, inplace = True)
operations.drop(operations.columns[[2,4,5,6]], axis = 1, inplace = True)
operations
```

#### Analysing appropriate airlines for traveling between the source and destination
**Process for same aircraft type between a city pair:**

**Analysis on verifying if the aircraft's range is greater than the distance between the source and destination:**

   **> If the condition is true, the cost of the trip is calculated as follows:**
   
   * Total cost of the trip = Cost per seat per km(USD) x Number of Seats x Distance(km) x Number of trips 

* Where, Number of trips = Pass. Demand (per day) / Number of Seats

```python
city_pairs
```

```python
operations
```

```python
import numpy as np
for a,b,c,d in city_pairs.itertuples(index=False):
    operations[str(a)+'-'+str(b)+' cost'] = np.where(operations['Range (Km)'] >= d, (operations['Cost per seat per km(USD)'] * operations['Number of Seats'] * d * np.ceil(c / operations['Number of Seats'])).astype(int), "NaN")

# Indicating the dataframe with added travel cost
operations
```

The following table summarizes the optimal aircraft type for each city pair along with the number of trips:



```python
indicies = ['AA-BB','BB-CC','CC-AA','AA-DD'] 
cols = ['Optimal aircraft type','Number of trips required']

# No. of trips = Pass demand per day / No. of seats
new_data = [['A330', np.ceil(420/250)], ['A330', np.ceil(450/250)], ['A320', np.ceil(300/150)], ['A320', np.ceil(300/150)]]
res = pd.DataFrame(new_data, index = indicies, columns = cols)
res
```

**Exploration for a mixture of aircraft type between a city pair:**

**In this case, the optimal aircraft types are found manually for each city pair.**

**City Pair AA-BB**



```python
df3 = operations.iloc[:,:4].copy()

# Travel cost and remaining passengers after 1st flight
df3['AA-BB cost 1'] = np.where(df3['Range (Km)'] >= 3000, (df3['Cost per seat per km(USD)'] * df3['Number of Seats'] * 3000).astype(int), 0)
df3['AA-BB remain 1'] = np.where(np.logical_and(420 - df3['Number of Seats'] >= 0, df3['AA-BB cost 1'] != 0), 420 - df3['Number of Seats'], 0)

# Total travel cost and remaining passengers after 2nd flight
df3['AA-BB cost 2'] = np.where(df3['Range (Km)'] >= 3000, (df3['Cost per seat per km(USD)'] * df3['Number of Seats'] * 3000).astype(int) + df3['AA-BB cost 1'], 0)
df3['AA-BB remain 2'] = np.where((df3['AA-BB remain 1'] - df3['Number of Seats']) >= 0, df3['AA-BB remain 1'] - df3['Number of Seats'], 0)
df3
```

**Inference for AA-BB: 2 aircrafts of type A330 are feasible.**



**City Pair BB-CC**

```python
df4 = operations.iloc[:,:4].copy()

# Travel cost and remaining passengers after 1st flight
df4['BB-CC cost 1'] = np.where(df4['Range (Km)'] >= 6500, (df4['Cost per seat per km(USD)'] * df4['Number of Seats'] * 6500).astype(int), 0)
df4['BB-CC remain 1'] = np.where(np.logical_and(450 - df4['Number of Seats'] >= 0, df4['BB-CC cost 1'] != 0), 450 - df4['Number of Seats'], 0)

# Total travel cost and remaining passengers after 2nd flight
df4['BB-CC cost 2'] = np.where(df4['Range (Km)'] >= 6500, (df4['Cost per seat per km(USD)'] * df4['Number of Seats'] * 6500).astype(int) + df4['BB-CC cost 1'], 0)
df4['BB-CC remain 2'] = np.where((df4['BB-CC remain 1'] - df4['Number of Seats']) >= 0, df4['BB-CC remain 1'] - df4['Number of Seats'], 0)
df4
```

**Inference for BB-CC: Since most of the aircrafts are not compatible for this journey, 2 aircrafts of type A330 are feasible.**



**City Pair CC-AA**

```python
df5 = operations.iloc[:,:4].copy()

# Travel cost and remaining passengers after 1st flight
df5['CC-AA cost 1'] = np.where(df5['Range (Km)'] >= 400, (df5['Cost per seat per km(USD)'] * df5['Number of Seats'] * 400).astype(int), 0)
df5['CC-AA remain 1'] = np.where(np.logical_and(300 - df5['Number of Seats'] >= 0, df5['CC-AA cost 1'] != 0), 300 - df5['Number of Seats'], 0)

# Total travel cost and remaining passengers after 2nd flight
df5['CC-AA cost 2'] = np.where(np.logical_and(df5['Range (Km)'] >= 400, df5['CC-AA remain 1'] != 0), (df5['Cost per seat per km(USD)'] * df5['Number of Seats'] * 400).astype(int) + df5['CC-AA cost 1'], df5['CC-AA cost 1'])
df5['CC-AA remain 2'] = np.where((df5['CC-AA remain 1'] - df5['Number of Seats']) >= 0, df5['CC-AA remain 1'] - df5['Number of Seats'], 0)
df5
```

**Inference for CC-AA: If 1 A330 and 1 ATR72 are used, the cost is still 25 Dollars more than using 2      aircrafts of A320 type. Hence, the A320 is more suitable than A330 and ATR72.**

**City Pair AA-DD**



```python
df6 = operations.iloc[:,:4].copy()

# Travel cost and remaining passengers after 1st flight
df6['AA-DD cost 1'] = np.where(df6['Range (Km)'] >= 1000, (df6['Cost per seat per km(USD)'] * df6['Number of Seats'] * 1000).astype(int), 0)
df6['AA-DD remain 1'] = np.where(np.logical_and(300 - df6['Number of Seats'] >= 0, df6['AA-DD cost 1'] != 0), 300 - df6['Number of Seats'], 0)

# Total travel cost and remaining passengers after 2nd flight
df6['AA-DD cost 2'] = np.where(np.logical_and(df6['Range (Km)'] >= 1000, df6['AA-DD remain 1'] != 0), (df6['Cost per seat per km(USD)'] * df6['Number of Seats'] * 1000).astype(int) + df6['AA-DD cost 1'], df6['AA-DD cost 1'])
df6['AA-DD remain 2'] = np.where((df6['AA-DD remain 1'] - df6['Number of Seats']) >= 0, df6['AA-DD remain 1'] - df6['Number of Seats'], 0)
df6
```

**Inference for AA-DD: In continuance to the previous result, 2 aircrafts of type A320 will be well-matched than a mixture of non-identical types.**
